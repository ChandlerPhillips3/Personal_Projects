```python
def MMSimCalc(Team1, Team2):
    Score1=0
    Score2=0
    
    ###Rank:
    Score1=Score1 + (17-Team1[1])
    Score2=Score2 + (17-Team2[1])
    
    PPG1=Team1[2]
    PPG2=Team2[2]
    ####PPG
    if(PPG1>PPG2):
        HoldDiff=PPG1-PPG2
        if HoldDiff>9:
            HoldDiff=9
        randomMult=.50+random.uniform(0,.75)
        Score1=Score1+(randomMult*HoldDiff)
        #print(randomMult*HoldDiff)

    if(PPG1<PPG2):
        HoldDiff=PPG2-PPG1
        if HoldDiff>9:
            HoldDiff=9
        randomMult=.50+random.uniform(0,.75)
        Score2=Score2+(randomMult*HoldDiff)
        #print(randomMult)
        
    #####OPP PPG
    OPP1=Team1[3]
    OPP2=Team2[3]
    if(OPP1>OPP2):
        ####ADD to other score
        HoldDiff=abs(OPP2-OPP1)
        if HoldDiff>9:
            HoldDiff=9
        randomMult=.50+random.uniform(0,.75)
        Score2=Score2+(randomMult*HoldDiff)
        #print(randomMult*HoldDiff)

    if(OPP1<OPP2):
        ####ADD to other score
        HoldDiff=abs(OPP1-OPP2)
        if HoldDiff>9:
            HoldDiff=9
        randomMult=.50+random.uniform(0,.75)
        Score1=Score1+(randomMult*HoldDiff)
        #print(randomMult*HoldDiff)

    ###Wins
    WIN1=Team1[4]
    WIN2=Team2[4]
    if(WIN1==WIN2):
        Holder1=2+random.uniform(0,3)
        Holder2=2+random.uniform(0,3)
        Score1=Score1+ Holder1
        Score2=Score2+ Holder2
    if(WIN1>WIN2):
        HoldDiff=abs(WIN1-WIN2)
        if HoldDiff>5:
            HoldDiff=5
        randomMult=.80+random.uniform(0,.4)
        Score1=Score1+(randomMult*HoldDiff)
    if(WIN2>WIN1):
        HoldDiff=abs(WIN2-WIN1)
        if HoldDiff>5:
            HoldDiff=5
        randomMult=.80+random.uniform(0,.4)
        Score2=Score2+(randomMult*HoldDiff)
    
    ###Losses
#     LOSS1=Team1[5]
#     LOSS2=Team1[5]
#     if(LOSS1<LOSS2):
#         HoldDiff=abs(LOSS2-LOSS1)
#         if HoldDiff>3:
#             HoldDiff=3
#         randomMult=.85+random.uniform(0,.30)
#         Score1=Score1+(randomMult*HoldDiff)
    
#     if(LOSS2<LOSS1):
#         HoldDiff=abs(LOSS1-LOSS2)
#         if HoldDiff>3:
#             HoldDiff=3
#         randomMult=.85+random.uniform(0,.30)
#         Score2=Score2+(randomMult*HoldDiff)
    
    
    PointsTotal=Score1+Score2
    Prob=Score1/PointsTotal
    #print(Prob)
    #########SIMUlate Games:
#     data = {'C':20, 'C++':15, 'Java':30,
#         'Python':35}
#     courses = list(data.keys())
#     values = list(data.values())

    fig = plt.figure(figsize = (5, 5))
    
    xlist=(Team1[0],Team2[0],"total prob")
    ylist=(Prob,1-Prob,1)
    # creating the bar plot
    plt.bar(xlist,ylist, color ='green',
            width = .2,)
    plt.xlabel("Teams")
    plt.ylabel("From Bottom is prob for first team")
    plt.title("GAME PROB")
    plt.show()
    Winner=0
    for j in range(11):
        result=np.random.choice(["Win", "Loss"], p=[Prob, 1-Prob])
        if(result=="Win"):
            Winner=Winner+1
    if(Winner>=6):
        print(Team1[0]," Moves on")
        return Team1
    else:
        print(Team2[0]," Moves on")
        return Team2
    
```
